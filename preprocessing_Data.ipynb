```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('DSSVK27GDQPAN.xlsx',index_col=None)
df.head()
```

```python
df = df.drop(df.columns[10:15],axis=1)
df.head()
```

```python
df = df.drop(df.columns[0], axis=1)
```

```python
df.head(10)
```

```python
df.columns = ['Số lượng', 'Thứ tự lớp','Mã sinh viên', 'Họ', 'Tên', 'Ngày Sinh', 'Giới Tính', 'Nơi Sinh', 'Lớp sinh hoạt',]
df.head(10)
```

```python
df = df.drop(index = 0)
df.head(10)
```

```python
# Chuyển giá trị của cột đầu tiên thành số, giá trị không phải số sẽ thành NaN
df['Thứ tự lớp'] = pd.to_numeric(df['Thứ tự lớp'],errors='coerce')
df = df.dropna(subset=['Thứ tự lớp'])
df 
```

```python
df = df.drop(df.columns[0],axis=1)
df
```

```python
df['Ngày Sinh'] = pd.to_datetime(df['Ngày Sinh'])
df['Ngày Sinh'] = df['Ngày Sinh'].dt.strftime('%d-%m-%Y')
df
```

```python
# Loại những sinh viên K28 ra khỏi danh sách
condition = df['Lớp sinh hoạt'].str.startswith('K27')
df = df[condition]
df
```

```python
df.to_csv('Danhsach_k27.csv',index=False)
```
