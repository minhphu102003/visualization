```python
track_artist = tiktok.groupby(['track_name'])['artist_name'].value_counts().unstack()
track_artist.head(10)
```

```python
artist_count = track_artist.sum(axis=0).dropna()
artist_count.head()
```

```python
artist_count = artist_count[artist_count > 0]
artist_count.head()
```

```python
top_artist = artist_count.sort_values(ascending=False).head(10)
top_artist
```

```python
top_artist.plot(kind='barh', color='brown')
plt.title('Top 10 Artists by Number of Songs')
plt.xlabel('Number of Songs')
plt.ylabel('Artist Name')
plt.show()
```

```python
artist_album = tiktok.groupby(['artist_name'])['album'].value_counts().unstack()
artist_album.head(10)
```

```python
artist_album_real = artist_album.sum(axis=0).dropna()
artist_album_real = artist_album_real[artist_album_real>0]
artist_album_real.head()
```

```python
top_artists = artist_album_real.sort_values(ascending=False).head(10)
top_artists
```

```python
top_artists.plot(kind='pie', autopct='%1.1F%%')
plt.title('Top 10 Artists by Album of Songs')
plt.show()
```
